```python
import json
from pathlib import Path


def read_jsonl(file_path):
    try:
        with open(file_path, "r", encoding="utf-8") as f:
            for line in f:
                yield json.loads(line)
    except GeneratorExit:
        print("Generator was closed")
```

```python
import re
import json
from collections import Counter, defaultdict

f = read_jsonl("/home/hadoop-mtai/users/zhangxiaoyun15/program/voc_explore/.tmp/DeepSeek-R1-Distill-Qwen-14B/traindata_stage_100.json")

cnt = Counter()
data_map = defaultdict(list)

def extract_answer_text(text):
    pattern = r'<answer>(.*?)</answer>'
    matches = re.findall(pattern, text, re.DOTALL)  # re.DOTALL 允许 . 匹配换行符
    return matches

for i, line in enumerate(f):
    target = line["target"]
    generated = line["generated"]
    count = 0
    for g in generated:
       if len(g) > 1000:
          print(g)
          print("---------")
       answers = extract_answer_text(g)
       if len(answers) == 0:
           break
       answer = answers[0]
       if answer == target:
          count += 1
    
    cnt[count] += 1
    data_map[count].extend([len(g) for g in generated])

for k, v in data_map.items():
    print(k, sum(v) / len(v))

```

```python
import matplotlib.pyplot as plt

sorted_items = sorted(data_map.items(), key=lambda x: x[0])
keys = [item[0] for item in sorted_items]
values = [sum(item[1]) / len(item[1]) for item in sorted_items]
values = [max(item[1]) for item in sorted_items]


# 绘制柱形图
plt.figure(figsize=(10, 6))
plt.bar(keys, values, color='skyblue')

plt.ylim(min(values) - 30, max(values) + 30)  # 上限比最大值大一些，以便显示数值标签

plt.xlabel('Key (k)')
plt.ylabel('平均值')
plt.title('按键排序的平均值柱形图')
plt.xticks(keys)  # 确保x轴刻度为整数键
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 显示每个柱子的数值
for i, v in enumerate(values):
    plt.text(keys[i], v + max(values)*0.01, f"{v:.2f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
